```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
import os

os.environ["PATH"] += os.pathsep + '/home/kasia/Pobrane/geckodrive/'

options = Options()
options.add_argument('--headless')

# Using Chrome to access web
driver = webdriver.Firefox(options=options)

website = 'http://www.azair.eu'

# Open the website
driver.get(website)
```

```python
from_box = driver.find_element_by_name('srcAirport')

driver.execute_script("arguments[0].value = ''", from_box)
from_box.send_keys(Keys.ARROW_DOWN)
driver.execute_script("arguments[0].value = ''", from_box)
```

```python
from_box.send_keys('France')
from_box.send_keys(Keys.ARROW_DOWN)
from_box.send_keys(Keys.RETURN)
```

```python
to_box = driver.find_element_by_name('dstAirport')

driver.execute_script("arguments[0].value = ''", to_box)
to_box.send_keys(Keys.ARROW_DOWN)
driver.execute_script("arguments[0].value = ''", to_box)

to_box.send_keys('XXX')
to_box.send_keys(Keys.ARROW_DOWN)
to_box.send_keys(Keys.RETURN)
```
